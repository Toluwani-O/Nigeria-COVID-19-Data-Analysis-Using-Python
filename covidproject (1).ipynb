Determine the effect of the Pandemic on the economy. To do this, you will compare the Real GDP value Pre-COVID-19 with Real GDP in 2020 (COVID-19 Period, especially Q2 2020)

Steps
* From the Real GDP Data, generate a `barplot` using the GDP values for each year & quarters. For example: On x-axis you will have year 2017 and the bars will be values of each quarters(Q1-Q4). You expected to have subplots of each quarters on one graph.

Hint: Use [Pandas.melt](https://pandas.pydata.org/docs/reference/api/pandas.melt.html) to create your plot DataFrame

```python
# Melt the DataFrame to create a suitable format for the barplot
melted_gdp = pd.melt(RealGDP, id_vars=['Year'], var_name='Quarter', value_name='GDP')

# Create subplots for each quarter
fig, ax = plt.subplots(figsize=(12, 5))
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
colors = ['b', 'g', 'r', 'c']

for i, quarter in enumerate(quarters):
    quarter_data = melted_gdp[melted_gdp['Quarter'] == quarter]
    ax.bar(quarter_data['Year'], quarter_data['GDP'], label=quarter, color=colors[i])

# Set labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Real GDP')
ax.set_title('Real GDP by Year and Quarter')
ax.legend(loc='lower left')

# Draw a horizontal line at the value of Q2 2020
q2_2020_gdp = RealGDP.loc[RealGDP['Year'] == 2020, 'Q2'].values[0]
ax.axhline(q2_2020_gdp, color='k', linestyle='--', label='Q2 2020')

# Show the plot
plt.tight_layout()
plt.show()
```
